# apple dataset work 

```python
#import what we need to start working on the data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# take a first look and try to understand the data

```python
#show the data
df = pd.read_csv('data/AppleStore.csv')
df.head()
```

```python
#remove the unwanted column
df = df.drop('Unnamed: 0', 1)
df.head()
```

## use the info() and describe() functions to get information about the data 

```python
df.info()
```

```python
df.describe()
```

```python
#check for null values
df.isnull().sum()
```

## matplotlib graphs

```python
#show the 20 apps with the best user rating
fig, axes = plt.subplots()
y = df.sort_values(['user_rating_ver'], ascending=False)['user_rating_ver'][:20]
x = df.sort_values(['user_rating_ver'], ascending=False)['track_name'][:20]
plt.xticks(rotation=90, fontsize=10)
plt.ylabel("user rating", fontsize = 15, rotation=90)
plt.title("The user rating of the top 20 rating", fontsize = 20)
axes.plot(x,y, 'g*-')
axes.set_ylim(0,5.1) 
```

### there is one app with a long name that ruin the visualization.
### therefore we will make it shorter 

```python
#search for this app's index
df.sort_values(['user_rating_ver'], ascending=False)['track_name'][:20]
```

```python
#set the name for a shorter one
df['track_name'][2972] = "Sunny"
```

```python
#now we will check again the list
df.sort_values(['user_rating_ver'], ascending=False)['track_name'][:20]
```

```python
# now lets plot the same graph with the new name
# we can see that the graph is more user friendly 
fig, axes = plt.subplots()
y = df.sort_values(['user_rating_ver'], ascending=False)['user_rating_ver'][:20]
x = df.sort_values(['user_rating_ver'], ascending=False)['track_name'][:20]
plt.xticks(rotation=90, fontsize=10)
plt.ylabel("user rating", fontsize = 15, rotation=90)
plt.title("The user rating of the top 20 rating", fontsize = 20)
axes.plot(x,y, 'g*-')
axes.set_ylim(0,5.1) 
plt.savefig("applePic/graph1.png", bbox_inches='tight')
```

```python
# barh plot which shows the 15 apps with the highest amount of rating
fig = plt.figure()
axes = fig.add_axes([0.1 ,0.1 ,0.9 ,0.9])
y = df.sort_values(['rating_count_tot'], ascending=False)['rating_count_tot'][:15]
x = df.sort_values(['rating_count_tot'], ascending=False)['track_name'][:15]

plt.xlabel("Amount of rating", fontsize = 20)
plt.ylabel("Track name", fontsize = 20, rotation = 90)
plt.title("Amount of rating of the top 15", fontsize = 20)
axes.barh(x,y)
plt.savefig("applePic/graph2.png", bbox_inches='tight')
```

```python
# from Personal knowledge we know that games are a big part of the app
# so we will check the ratio between games and non-games apps
# Pie chart which shows us the ratio between the amount of games and the amount of non-games
labels = 'Games', 'Others'
gamesAmount = df[df['prime_genre'] == "Games"]['prime_genre'].count()
othersAmount = df[df['prime_genre'] != "Games"]['prime_genre'].count()
sizes = [gamesAmount, othersAmount]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title("The ratio between the Games and the other Prime genre", fontsize = 20)

plt.savefig("applePic/graph3.png", bbox_inches='tight')
```

```python

# bar plot to show us the average mega byte for each prime genre
fig = plt.figure()
axes = fig.add_axes([0.1 ,0.1 ,0.9 ,0.9])
y = ((df.groupby('prime_genre')['size_bytes'].mean())/1e6)
x = df['prime_genre'].unique()
plt.xticks(rotation=90, fontsize=10)
plt.xlabel("prime genre", fontsize = 20)
plt.ylabel("mega byte", fontsize = 20)
plt.xticks(fontsize = 13)
plt.title("The average of mega byte for each prime genre", fontsize = 20)
plt.bar(x,y)

plt.savefig("applePic/graph4.png", bbox_inches='tight')
```

```python
# now lets plot 2 graph in the same figure

# first graph (axes) will show us the the ratio between support devices and support language for each app
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (11,15))
axes[0].scatter(df['sup_devices.num'],df['lang.num'])
axes[0].set_xlabel('Number of supported devices', fontsize = 15)
axes[0].set_ylabel('Number of supported languages', fontsize = 15)
axes[0].set_title('The ratio between support devices and support language for each app', fontsize = 20)

# move to the second graph
plt.subplot(2,1,2)
# plot a scatter graph that shows the number of supported languages for each user rating 
y = df['user_rating']
x = df['lang.num']
plt.scatter(y,x, marker = 'X')
plt.xlabel('user rating', fontsize = 15)
plt.ylabel('number of support languages', fontsize = 15)
plt.title('The number of supported languages for each user rating', fontsize = 20)
fig.tight_layout()

plt.savefig("applePic/graph5and6.png", bbox_inches='tight')
```

## seaborn graphs

```python
#import seaborn by the name sns 
import seaborn as sns
```

```python
# creat a copy of df which prime genre column is only games
dfGames = df[df['prime_genre'] == 'Games'].copy()
dfGames.head()
```

```python
# creat a new series of boolean values to show if the game is free or not and add it to dfgames
ser = dfGames['price'] == 0
dfGames['Free'] = ser 
dfGames.head(10)
```

```python
# A pairplot for all the free games
dfFreeGames = dfGames[dfGames['Free'] == True]
sns.pairplot(dfFreeGames)
plt.savefig("applePic/graph7.png", bbox_inches='tight')
```

```python
#pair plot for all the non-free games
dfPaidGames = dfGames[dfGames['Free'] == False]
sns.pairplot(dfPaidGames)
plt.savefig("applePic/graph8.png", bbox_inches='tight')
```

```python
# creat a new series of boolean values to show if the app is free or not (same as the games but with all the apps this time)
# and add it to df
ser = df['price'] == 0
df['free'] = ser
df.head()
```

```python
# earlier we saw the ratio between the game and non-games apps
# so now lets countplot of all the unique values in prime_genre and see the difference between them 
fig = plt.figure(figsize=(34,17))
axes = fig.add_axes([0.1 ,0.1 ,0.9 ,0.9])
yticks = np.linspace(0,4000,21)
axes.set_yticks(yticks)
plt.yticks(fontsize= 30)
plt.xticks(rotation=90, fontsize=30)
plt.xlabel("", fontsize = 50)
plt.ylabel("", fontsize = 50)
plt.title("The amount of each prime genre", fontsize = 60)
sns.countplot(x="prime_genre" ,data=df, ax = axes)
plt.savefig("applePic/graph9.png", bbox_inches='tight')
```

```python
# violinplot to see the difference between the free and non-free apps in the user rating section 
sns.violinplot(x="free", y="user_rating", data=df)
plt.savefig("applePic/graph10.png", bbox_inches='tight')
```

```python
# stripplot to see the number of supported devices for each app in the prime genre
fig = plt.figure(figsize = (7,7))
axes = fig.add_axes([0.1 ,0.1 ,0.9 ,0.9])
plt.yticks(fontsize= 10)
plt.xticks(rotation=90, fontsize=13)
plt.xlabel("", fontsize = 17)
plt.ylabel("", fontsize = 17)
plt.title("Number of supported devices for each prime genre", fontsize = 20)
sns.stripplot(x="prime_genre", y="sup_devices.num", data=df)
plt.savefig("applePic/graph11.png", bbox_inches='tight')
```

```python
# violinplot to see the number of supported languages for each app in the prime genre
# and the difference between the free and non-free apps
fig = plt.figure(figsize=(10,6))
dftemp = df.copy()
for x in range(dftemp.shape[0]):
    if(dftemp['prime_genre'][x] != 'Games'):
        dftemp['prime_genre'][x] = 'Others'
sns.violinplot(x="prime_genre", y="lang.num",data=dftemp ,hue='free', split=True)
plt.savefig("applePic/graph12.png", bbox_inches='tight')
```
