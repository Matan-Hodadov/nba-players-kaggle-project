```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/AppleStore.csv')
df.head()
```

```python
df = df.drop('Unnamed: 0', 1)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
fig, axes = plt.subplots()
y = df.sort_values(['user_rating_ver'], ascending=False)['user_rating_ver'][:20]
x = df.sort_values(['user_rating_ver'], ascending=False)['track_name'][:20]
plt.xticks(rotation=90, fontsize=10)
axes.plot(x,y, 'g*-')
axes.set_ylim(0,5.1) 
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1 ,0.1 ,0.9 ,0.9])
y = df.sort_values(['rating_count_tot'], ascending=False)['rating_count_tot'][:15]
x = df.sort_values(['rating_count_tot'], ascending=False)['track_name'][:15]
axes.barh(x,y)
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Games', 'Others'
gamesAmount = df[df['prime_genre'] == "Games"]['prime_genre'].count()
othersAmount = df[df['prime_genre'] != "Games"]['prime_genre'].count()
sizes = [gamesAmount, othersAmount]
#explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs') , explode=explode

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1 ,0.1 ,0.9 ,0.9])
y = df['user_rating']
x = df['lang.num']
axes.scatter(y,x, marker = 'X')
axes.set_xlabel('user rating', fontsize = 17)
axes.set_ylabel('number of support languages', fontsize = 17)
axes.set_title('Plot_Title')
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8,13))
axes[0].scatter(df['sup_devices.num'],df['lang.num'])
axes[0].set_xlabel('support devices')
axes[0].set_ylabel('support language')
axes[0].set_title('the ratio between support devices app and support language')

plt.subplot(2,1,2)
y = ((df['size_bytes'])/1e6)
x = df['prime_genre']
plt.xticks(rotation=90, fontsize=10)
plt.xlabel("prime genre", fontsize = 20)
plt.ylabel("mega byte", fontsize = 20)
plt.bar(x,y)
```

```python
import seaborn as sns
```

```python
dfGames = df[df['prime_genre'] == 'Games'].copy()
dfGames.head()
```

```python
# ser = dfGames[dfGames['price'] == 0.00]['price']
ser = dfGames['price'] == 0
dfGames['Free'] = ser 
dfGames.head(10)
```

```python
dfFreeGames = dfGames[dfGames['Free'] == True]
sns.pairplot(dfFreeGames)
```

```python
dfPaidGames = dfGames[dfGames['Free'] == False]
sns.pairplot(dfPaidGames)
```

```python

ser = df['price'] == 0
df['free'] = ser
df.head()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1 ,0.1 ,0.9 ,0.9])
plt.yticks(fontsize= 10)
plt.xticks(rotation=90, fontsize=10)
plt.xlabel("", fontsize = 20)
plt.ylabel("", fontsize = 20)
sns.stripplot(x="prime_genre", y="sup_devices.num", data=df)
plt.show()
```

```python
sns.violinplot(x="free", y="user_rating", data=df)
```

```python
fig = plt.figure(figsize=(10,6))
dftemp = df.copy()
for x in range(dftemp.shape[0]):
    if(dftemp['prime_genre'][x] != 'Games'):
        dftemp['prime_genre'][x] = 'Others'
sns.violinplot(x="prime_genre", y="lang.num",data=dftemp ,hue='free', split=True)
```

```python
fig = plt.figure(figsize=(34,17))
axes = fig.add_axes([0.1 ,0.1 ,0.9 ,0.9])
yticks = np.linspace(0,4000,21)
axes.set_yticks(yticks)
plt.yticks(fontsize= 20)
plt.xticks(rotation=90, fontsize=30)
plt.xlabel("", fontsize = 50)
plt.ylabel("", fontsize = 50)
sns.countplot(x="prime_genre" ,data=df, ax = axes)
plt.show()
```
